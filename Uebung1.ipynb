```python
# Import some necessary packages
import numpy,scipy
import matplotlib.pyplot as p
```

## Part 1
Plotting the 1-D Gaussian-Function
$$ y(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}} $$

```python
def gaussian_1D(x,sigma,mu):
    y = 1/numpy.sqrt(2*numpy.pi*sigma**2) * numpy.exp(-(x-mu)**2/(2*sigma**2))
    return y
```

```python
N = 200 # numbers of data points
x = numpy.linspace(-5,5,N)

sigma = 1 # parameters
mu = 1
y = gaussian_1D(x,sigma,mu)

Graph1 = p.figure(dpi=100)
Graph1.patch.set_edgecolor('none')
p.plot(x,y,'r')

# Titles and labels
p.title("Plot a 1-D Gaussian-function")
p.xlabel("x")
p.ylabel("y")
p.show()
```

## Part 2
Plotting the 2-D Gaussian-Function
$$ y(x_1,x_2) = \frac{1}{\sqrt{2\pi\sigma_1\sigma_2}} e^{-\frac{(x_1-\mu_1)^2}{2 \sigma_1^2} -\frac{(x_2-\mu_2)^2}{2 \sigma_1^2}} $$

```python
from mpl_toolkits.mplot3d import Axes3D

def gaussian_2D(x, y, x_0=0, y_0=0, sigma=1):
    return 1/numpy.sqrt(2*numpy.pi*sigma**2) * numpy.exp(-(x-x_0)**2/(2*sigma**2)-(y-y_0)**2/(2*sigma**2))
```

```python
N = 100 # numbers of data points
x_Data = numpy.linspace(-5,5,N)
x,y = numpy.meshgrid(x_Data,x_Data)
# y,x = numpy.mgrid[-5:5:100j,-5:5:100j]

sigma = 1 # parameters
x_0 = 0
y_0 = 0
z = gaussian_2D(x,y,x_0,y_0,sigma)

Graph2 = p.figure(dpi=100)
ax = p.axes(projection='3d') # Add an axes to the current figure and make it the current axes.
# ax = Graph2.add_axes(projection='3d')
# ax = Graph2.add_subplot(1,1,1,projection='3d')

ax.plot_surface(x,y,z,cmap='viridis',edgecolor='none');

# Titles and labels
p.title("Plot a 2-D Gaussian-function")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
p.show()
```
