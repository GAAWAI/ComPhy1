```python
import numpy,scipy
import matplotlib.pyplot as p
```

UE 1 CP 1

```python
sigma = 1
mu = 1
N = 100 #numbers of data points
x = numpy.linspace(-5,5,N)
y = 1/numpy.sqrt(2*numpy.pi*sigma**2) * numpy.exp(-(x-mu)**2/(2*sigma**2))
GRAPH1 = p.plot(x,y,'r.')

# Titles and labels
p.title("Plot a 1-D Gaussian-function")
p.xlabel("x")
p.ylabel("y")
p.show(GRAPH1)
```

Plotting the 1-D Gaussian-Function $$ y(x,\sigma,\mu) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2 \sigma^2}} $$

```python
from mpl_toolkits.mplot3d import Axes3D

def gaussian2(x,y,x_0,y_0,sigma):
    return 1/numpy.sqrt(2*numpy.pi*sigma**2)  * numpy.exp(-(x-x_0)**2/(2*sigma**2)-(y-y_0)**2/(2*sigma**2))

FIG = p.figure(100)
ax = p.axes(projection='3d')

sigma = 1
x_0 = 0
y_0 = 0
N = 100 #numbers of data points

x = numpy.linspace(-5,5,N)
y = numpy.linspace(-5,5,N)
x,y = numpy.meshgrid(x,y)
z2 = gaussian2(x,y,x_0,y_0,sigma)

GRAPH1 = ax.plot_surface(x,y,z2,cmap='viridis',edgecolor='none');

# Titles and labels
p.title("Plot a 2-D Gaussian-function")
p.xlabel("x")
p.ylabel("y")
p.show(GRAPH1)
```
